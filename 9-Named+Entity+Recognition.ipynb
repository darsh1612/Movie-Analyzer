### Named Entity Recognition

```python
sentence="The Eiffel Tower And Taj Mahal was built from 1887 to 1889 by Gustave Eiffel And akbar, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
words=nltk.word_tokenize(sentence)
print(words)
```

```python
tag_elements=nltk.pos_tag(words)
print(tag_elements)
```
