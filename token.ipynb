```python
corpus = '''Hello, how are you? I hope you are doing well. 
 Have a great day! I am here to help you with your questions.
 Also I can assist you with various tasks such as writing,
 brainstorming, and providing information on a wide range of topics.'''

print(corpus)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokenize(corpus)
```

```python
words = word_tokenize(corpus)
for i in words:
    print(i)
```
